### Checking for missing values

```python
data.dropna(axis="index", how="any", inplace=True)
```

### Checking for incoherent values

```python
dataframe_85 = data[["fatal_accidents_85_99", "fatalities_85_99"]]
dataframe_00 = data[["fatal_accidents_85_99", "fatalities_85_99"]]
```

```python
def check_incoherence(dataframe):
    for index, row in dataframe.iterrows():
        fatal_accidents, fatalities = row
        if (fatal_accidents != 0 and fatalities == 0) or (fatal_accidents == 0 and fatalities != 0):
            print(f"Incoherence detected in line: {index}")
```

```python
check_incoherence(dataframe_85)
check_incoherence(dataframe_00)
```

#### The data seems to be coherent in terms of casualities

### Questions.

Is the number of incidents correlated with the avail_seat_km/week ?

are the nb of fatal accidents and incidents correlated ?

Did the airlines get safer ?


```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
correlation_matrix = data.corr( )
sn.heatmap(correlation_matrix, annot=True)
plt.show()
```

- The incidents and fatal incidents seem correlated, especially in the first period from 85 to 99. This is however less true in the period from 00 to 14.

- We also see that the avail_seat_km_per_week is quite correlated with the incidents of 00-14, but not that much with those of 85-99, which seems counter-intuitive. One would expect that companies get safer over time as they gain more experience from their past mistakes. This might be due to the fact that the "incidents" from 85-99 tend to be way deadlier, becoming fatal_accidents


#### Verifying the total number of incidents of each period.

```python
summed_data = data.iloc[:, 2:8].sum(axis = 0)

labels = ["85-99 period",
          "00-14 period"]
x = [1, 2]
incidents = [summed_data.incidents_85_99, summed_data.incidents_00_14]
fatalities = [summed_data.fatalities_85_99, summed_data.fatalities_00_14]
fatal_accidents = [summed_data.fatal_accidents_85_99, summed_data.fatal_accidents_00_14]
```

```python
ax = plt.subplot(111)
ax.bar(x, incidents, width=0.2, color='b', align='center')
ax.bar(list(map(lambda x: x - 0.2, x)), fatalities, width=0.2, color='g', align='center')
ax.bar(list(map(lambda x: x + 0.2, x)), fatal_accidents, width=0.2, color='r', align='center')
plt.xticks([1, 2], labels)
plt.legend(["incidents", "fatalities", "fatal_accidents"])
plt.ylabel("Frequency")
plt.title("Comparison of nb. of occurences between the two periods.")
plt.plot()
```

- We can see that the airlines got safer in all 3 categories.

- We however still need to see the ratios of casualties/fatal_accidents, and incidents/fatal_accidents
